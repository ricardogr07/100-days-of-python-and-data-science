# **Day 27 — Visualizing Data with Pie Charts and Donut Charts**

### **Why Use Pie Charts and Donut Charts?**

Pie charts and donut charts are effective at displaying data where you want to highlight the proportions of categories relative to the whole. These visualizations are especially useful for:
- **Showing Part-to-Whole Relationships**: Quickly see how individual categories contribute to the total.
- **Comparing Segments**: Visually compare the size of different segments.
- **Highlighting Key Data**: Emphasize specific segments to draw attention to important data points.

While pie charts and donut charts can be impactful, they should be used wisely—too many slices can make them hard to read. For clear and compelling insights, limit your data to a manageable number of categories.



### **Creating and Customizing Pie Charts in Python**

We will use Python’s Matplotlib library to create and customize pie charts and donut charts. Matplotlib offers versatile options for enhancing your charts with colors, labels, and annotations.

#### **Step 1: Importing the Necessary Libraries**

Ensure you have Matplotlib installed. If not, you can install it using the command `pip install matplotlib`.



```python

import matplotlib.pyplot as plt
```


#### **Step 2: Creating a Basic Pie Chart**

Pie charts represent data as slices of a circle, where each slice corresponds to a category's contribution to the total.


```python

# Sample Data: Market share of different companies
labels = ['Company A', 'Company B', 'Company C', 'Company D']
sizes = [35, 30, 20, 15]  # Market share percentages

# Creating a basic pie chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Market Share Distribution')
plt.show()
```


#### **Step 3: Customizing Pie Charts**

Customizing your pie charts with colors, explode effects, and annotations can enhance readability and emphasize key data points.


```python

# Customizing the pie chart with colors and explode effect
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']
explode = (0.1, 0, 0, 0)  # Explode the first slice (Company A)

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors, explode=explode, shadow=True)
plt.title('Customized Market Share Distribution')
plt.show()
```


#### **Step 4: Creating Donut Charts**

Donut charts are essentially pie charts with a hollow center, providing a modern and stylish alternative that can also display additional information in the center.


```python

# Creating a basic donut chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors)
# Adding a white circle in the middle to create the donut shape
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
plt.gca().add_artist(centre_circle)
plt.title('Market Share Donut Chart')
plt.show()
```


### **Use Case: Visualizing Market Share Data**

Let’s apply these techniques to visualize market share data for a group of companies in a specific industry. This scenario is common in business analysis, where understanding the market dynamics is key to strategic decision-making.

#### **Step 1: Simulating Market Share Data**

We will create a dataset representing the market share of different companies within a tech industry segment.


```python

# Simulated market share data for a tech industry segment
company_labels = ['Tech Corp', 'Innovate Ltd', 'Future Tech', 'Alpha Inc', 'Beta Systems']
market_shares = [40, 25, 15, 10, 10]  # Percentages

# Plotting the market share as a pie chart
plt.figure(figsize=(10, 10))
plt.pie(market_shares, labels=company_labels, autopct='%1.1f%%', startangle=140, colors=plt.cm.Pastel1.colors)
plt.title('Tech Industry Market Share')
plt.show()
```


#### **Step 2: Customizing for Greater Insight**

We can customize this chart further to draw attention to the market leader.


```python

# Highlighting the market leader with an exploded slice
explode = (0.1, 0, 0, 0, 0)  # Explode the first slice (Tech Corp)

plt.figure(figsize=(10, 10))
plt.pie(market_shares, labels=company_labels, autopct='%1.1f%%', startangle=140, colors=plt.cm.Pastel1.colors, explode=explode, shadow=True)
plt.title('Tech Industry Market Share with Highlighted Leader')
plt.show()
```


**Key Takeaways**:
- **Pie Charts** are effective for showing how individual parts contribute to a whole, with customization options to enhance impact.
- **Donut Charts** provide a modern twist on pie charts, adding visual appeal while maintaining clarity.
- **Customizations** such as colors, exploded slices, and shadows can draw attention to key segments and improve chart readability.
