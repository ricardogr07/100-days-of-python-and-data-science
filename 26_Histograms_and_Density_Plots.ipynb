# **Day 26 — Visualizing Histograms and Density Plots in Python**

### **Why Are Histograms and Density Plots Important?**

Histograms and density plots provide a graphical representation of data distributions, helping you quickly identify the shape, spread, and central tendencies of your data. They are particularly useful for:
- **Understanding Data Distribution**: Quickly see how your data is spread across different values, including detecting skewness, peaks, or gaps.
- **Identifying Outliers**: Spot unusual data points that fall outside the general pattern of your dataset.
- **Comparing Distributions**: Visualize multiple datasets to compare their distributions side-by-side.



### **Creating Histograms and Density Plots in Python**

We’ll use Python’s popular libraries, Matplotlib and Seaborn, to create histograms and density plots. Seaborn builds on Matplotlib to provide a high-level interface for drawing attractive statistical graphics.

#### **Step 1: Importing the Necessary Libraries**

Then, import the necessary libraries:


````python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
````


#### **Step 2: Creating a Basic Histogram**

Histograms are great for visualizing the distribution of numerical data by dividing the data into bins and counting the number of data points that fall into each bin.


````python

# Sample Data: Simulated exam scores of 100 students
np.random.seed(42)
scores = np.random.normal(loc=70, scale=15, size=100)  # Mean=70, Std Dev=15

# Creating a DataFrame
df = pd.DataFrame({'Scores': scores})

# Plotting a basic histogram
plt.figure(figsize=(10, 6))
plt.hist(df['Scores'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Exam Scores')
plt.xlabel('Scores')
plt.ylabel('Frequency')
plt.grid(True, linestyle='--', linewidth=0.5)
plt.show()
````


#### **Step 3: Creating a Density Plot (KDE Plot)**

A Kernel Density Estimate (KDE) plot provides a smooth, continuous curve representing the distribution of data. It’s especially useful for visualizing the probability density function of a continuous variable.


````python

# Plotting a basic density plot
plt.figure(figsize=(10, 6))
sns.kdeplot(df['Scores'], shade=True, color='red')
plt.title('Density Plot of Exam Scores')
plt.xlabel('Scores')
plt.ylabel('Density')
plt.grid(True, linestyle='--', linewidth=0.5)
plt.show()
````


#### **Step 4: Combining Histograms and Density Plots**

Combining histograms and KDE plots provides a comprehensive view of the data distribution, highlighting both the frequency and density.


````python

# Combining Histogram and KDE Plot
plt.figure(figsize=(10, 6))
sns.histplot(df['Scores'], bins=10, kde=True, color='purple', edgecolor='black')
plt.title('Combined Histogram and Density Plot of Exam Scores')
plt.xlabel('Scores')
plt.ylabel('Frequency/Density')
plt.grid(True, linestyle='--', linewidth=0.5)
plt.show()
````


### **Use Case: Analyzing the Distribution of Exam Scores**

Let’s apply these techniques to a practical scenario—analyzing exam scores to understand student performance distribution and identify any patterns or anomalies.

#### **Step 1: Simulating Exam Score Data**

We’ll simulate exam score data to represent a typical classroom scenario, where scores follow a normal distribution with some students performing better or worse than the average.


````python

# Simulating exam scores of 200 students
np.random.seed(123)
exam_scores = np.random.normal(loc=75, scale=10, size=200)  # Mean=75, Std Dev=10

# Creating a DataFrame
exam_df = pd.DataFrame({'Exam Scores': exam_scores})
````


#### **Step 2: Visualizing the Exam Scores with a Histogram**

We start by plotting a histogram to visualize the spread of scores.


````python

# Plotting the histogram of exam scores
plt.figure(figsize=(12, 6))
sns.histplot(exam_df['Exam Scores'], bins=15, kde=False, color='blue', edgecolor='black')
plt.title('Histogram of Exam Scores')
plt.xlabel('Scores')
plt.ylabel('Frequency')
plt.grid(True, linestyle='--', linewidth=0.5)
plt.show()
````


#### **Step 3: Adding a Density Plot to Highlight Distribution Shape**

Next, we add a KDE plot to the histogram to highlight the shape and smoothness of the score distribution.


````python

# Histogram with KDE
plt.figure(figsize=(12, 6))
sns.histplot(exam_df['Exam Scores'], bins=15, kde=True, color='green', edgecolor='black')
plt.title('Histogram and Density Plot of Exam Scores')
plt.xlabel('Scores')
plt.ylabel('Frequency/Density')
plt.grid(True, linestyle='--', linewidth=0.5)
plt.show()
````


**Key Takeaways**:
- **Histograms** provide a quick way to visualize data distribution by showing the frequency of data points within specified ranges.
- **Density Plots (KDE)** offer a smooth, continuous visualization of the data distribution, highlighting density without being constrained by bins.
- **Combined Plots** allow you to see both the frequency and density, giving a comprehensive view of the data's underlying patterns.
