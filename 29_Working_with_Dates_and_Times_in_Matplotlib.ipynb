
# **Day 29 — Working with Dates and Times in Matplotlib**

### **Why Plot Time Series Data?**

Time series plots are powerful tools for tracking changes over time. They help you:
- **Identify Trends**: See long-term movements in data, such as increasing sales, declining temperatures, or seasonal patterns.
- **Spot Patterns and Anomalies**: Detect regular cycles, spikes, or unexpected events.
- **Forecast Future Values**: Use historical data to predict future outcomes, such as stock prices or weather conditions.

These insights are invaluable in fields like finance, meteorology, economics, and many other domains where understanding changes over time is key.



### **Plotting Time Series Data in Python**

Python’s Matplotlib and Pandas libraries provide robust functionality for handling dates and times in plots. In this tutorial, we’ll explore how to plot time series data using these libraries.

#### **Step 1: Importing the Necessary Libraries**

Ensure you have Matplotlib and Pandas installed. If not, you can install them using the command `pip install matplotlib pandas`.


```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```


#### **Step 2: Plotting Time Series Data**

We’ll start by creating a simple time series plot using random data to demonstrate how to handle date and time axes.


```python

# Sample Data: Generating a time series of random data
date_range = pd.date_range(start='2023-01-01', periods=100, freq='D')
data = pd.DataFrame({
    'Date': date_range,
    'Value': pd.Series(np.random.randn(100)).cumsum()
})

# Plotting the time series
plt.figure(figsize=(12, 6))
plt.plot(data['Date'], data['Value'], marker='o', linestyle='-')
plt.title('Sample Time Series Data')
plt.xlabel('Date')
plt.ylabel('Cumulative Value')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```


#### **Step 3: Formatting Date and Time Axes**

Customizing the date and time formatting on your plots can improve readability and make your visualizations more informative.


```python

# Formatting the date axis
plt.figure(figsize=(12, 6))
plt.plot(data['Date'], data['Value'], marker='o', linestyle='-', color='purple')
plt.title('Formatted Time Series Data')
plt.xlabel('Date')
plt.ylabel('Cumulative Value')
plt.grid(True)
plt.xticks(rotation=45)

# Formatting dates to show month and year
plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%b %Y'))

plt.tight_layout()
plt.show()
```


#### **Step 4: Plotting Multiple Time Series**

You can plot multiple time series on the same plot to compare trends, making it ideal for comparing datasets like sales across regions or temperatures across different cities.


```python

# Sample Data: Generating multiple time series
data['Value2'] = pd.Series(np.random.randn(100)).cumsum()
data['Value3'] = pd.Series(np.random.randn(100)).cumsum()

# Plotting multiple time series
plt.figure(figsize=(14, 7))
plt.plot(data['Date'], data['Value'], label='Series 1', marker='o')
plt.plot(data['Date'], data['Value2'], label='Series 2', marker='s')
plt.plot(data['Date'], data['Value3'], label='Series 3', marker='^')
plt.title('Multiple Time Series Comparison')
plt.xlabel('Date')
plt.ylabel('Cumulative Value')
plt.grid(True)
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```


### **Use Case: Visualizing Historical Temperature Trends**

Let’s apply these techniques to a practical scenario: visualizing historical temperature trends. This can help us understand how temperatures have changed over time and identify any significant patterns.

#### **Step 1: Loading Historical Temperature Data**

For this use case, we’ll simulate a dataset representing monthly average temperatures over several years.


```python

# Simulating historical temperature data
date_range = pd.date_range(start='2000-01-01', periods=240, freq='M')
temperatures = pd.Series(np.random.normal(15, 5, size=240))  # Average temperature with some randomness

# Creating a DataFrame
temperature_data = pd.DataFrame({
    'Date': date_range,
    'Temperature': temperatures
})
```


#### **Step 2: Plotting the Temperature Trends**

We’ll plot the temperature data to visualize how temperatures have varied over the years.


```python

# Plotting historical temperature trends
plt.figure(figsize=(14, 7))
plt.plot(temperature_data['Date'], temperature_data['Temperature'], marker='o', linestyle='-', color='orange')
plt.title('Historical Temperature Trends')
plt.xlabel('Year')
plt.ylabel('Average Temperature (°C)')
plt.grid(True)
plt.xticks(rotation=45)

# Formatting the x-axis to show the year
plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%Y'))

plt.tight_layout()
plt.show()
```
