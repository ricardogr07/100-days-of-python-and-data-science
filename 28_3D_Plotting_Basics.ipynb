# **Day 28 — 3D Plotting Basics in Python**

### **Why Use 3D Plots?**

3D plots are essential when dealing with multi-dimensional data. They provide a deeper insight into the data structure and relationships between variables, allowing you to:
- **Visualize Multi-Dimensional Data**: Easily see how different variables interact with one another.
- **Highlight Trends and Patterns**: Observe correlations or outliers that would be missed in 2D visualizations.
- **Enhance Presentations**: Make data-driven storytelling more compelling with dynamic, interactive visuals.



### **Generating Basic 3D Plots in Python**

Python’s Matplotlib library offers robust 3D plotting capabilities via its `mpl_toolkits.mplot3d` module. In this tutorial, we’ll walk through how to create basic 3D plots, including scatter, line, and surface plots.

#### **Step 1: Importing the Necessary Libraries**

Make sure you have Matplotlib installed. If not, you can install it using the command `pip install matplotlib`.



```python

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```


#### **Step 2: Creating a Basic 3D Scatter Plot**

3D scatter plots are great for visualizing the relationships between three variables, showing data points in three-dimensional space.


```python

# Generating random 3D data points
np.random.seed(42)
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)

# Creating a 3D scatter plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, c='r', marker='o')

# Setting labels
ax.set_title('3D Scatter Plot')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')

plt.show()
```


#### **Step 3: Creating a 3D Line Plot**

3D line plots are used to connect data points in three dimensions, which can help visualize paths or trends in multi-dimensional data.


```python

# Generating data for a 3D line plot
t = np.linspace(0, 20, 100)
x = np.sin(t)
y = np.cos(t)
z = t

# Creating a 3D line plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z, label='3D Line Plot', color='b')

# Setting labels
ax.set_title('3D Line Plot')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')
ax.legend()

plt.show()
```


#### **Step 4: Creating a 3D Surface Plot**

3D surface plots visualize data as a continuous surface, allowing you to observe how values change smoothly over a grid. They are particularly useful for visualizing mathematical functions or landscape data.


```python

# Generating data for a 3D surface plot
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Creating a 3D surface plot
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')

# Setting labels
ax.set_title('3D Surface Plot')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')

plt.show()
```


### **Use Case: Visualizing 3D Data Points**

Let’s apply 3D plotting techniques to a practical use case—visualizing 3D data points. This scenario could represent anything from geographical data points, sensor data, or complex relationships in financial datasets.

#### **Step 1: Generating 3D Data Points**

We will simulate a set of 3D data points representing an imaginary dataset of measurements taken over time.


```python

# Simulating 3D data points for visualization
np.random.seed(101)
x = np.random.normal(0, 1, 100)
y = np.random.normal(0, 1, 100)
z = np.random.normal(0, 1, 100)

# Creating a 3D scatter plot
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(x, y, z, c=z, cmap='coolwarm', s=50)

# Adding a color bar for the z axis values
fig.colorbar(scatter, ax=ax, shrink=0.5, aspect=5)

# Setting labels
ax.set_title('3D Data Point Visualization')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')

plt.show()
```
