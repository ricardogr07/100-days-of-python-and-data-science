```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Generating months
months = pd.date_range(start='2024-01-01', periods=12, freq='M').strftime('%b')

# Generating sales data for three products
np.random.seed(25)  # For reproducibility
product_a_sales = np.random.randint(20, 50, size=12)
product_b_sales = np.random.randint(30, 60, size=12)
product_c_sales = np.random.randint(10, 40, size=12)

# Creating a DataFrame
sales_data = pd.DataFrame({
    'Months': months,
    'Product A': product_a_sales,
    'Product B': product_b_sales,
    'Product C': product_c_sales
})

# Displaying the DataFrame
display(sales_data)
```

```python
# Setting the figure size
plt.figure(figsize=(12, 6))

# Plotting each product's sales
plt.plot(sales_data['Months'], sales_data['Product A'], marker='o', linestyle='-', label='Product A')
plt.plot(sales_data['Months'], sales_data['Product B'], marker='s', linestyle='--', label='Product B')
plt.plot(sales_data['Months'], sales_data['Product C'], marker='^', linestyle=':', label='Product C')

# Displaying the plot
plt.show()
```

```python
# Generating months
months = pd.date_range(start='2024-01-01', periods=12, freq='M').strftime('%b')

# Generating random weather data
np.random.seed(42)
temperature = np.random.uniform(low=10, high=35, size=12)
humidity = np.random.uniform(low=30, high=90, size=12)
precipitation = np.random.uniform(low=0, high=200, size=12)

# Creating a DataFrame
weather_data = pd.DataFrame({
    'Months': months,
    'Temperature (C)': temperature,
    'Humidity (%)': humidity,
    'Precipitation (mm)': precipitation
})

# Displaying the DataFrame
print(weather_data)
```

```python
# Creating figure and axis
fig, ax1 = plt.subplots(figsize=(14, 7))

# Plotting Temperature
ax1.plot(weather_data['Months'], weather_data['Temperature (C)'], color='tab:red', marker='o', label='Temperature')
ax1.set_xlabel('Months', fontsize=12)
ax1.set_ylabel('Temperature (C)', color='tab:red', fontsize=12)
ax1.tick_params(axis='y', labelcolor='tab:red')
ax1.set_title('Weather Statistics Over a Year', fontsize=16, fontweight='bold')

# Creating a twin axis for Humidity
ax2 = ax1.twinx()
ax2.plot(weather_data['Months'], weather_data['Humidity (%)'], color='tab:blue', marker='s', label='Humidity')
ax2.set_ylabel('Humidity (%)', color='tab:blue', fontsize=12)
ax2.tick_params(axis='y', labelcolor='tab:blue')

# Creating another twin axis for Precipitation
ax3 = ax1.twinx()
ax3.spines['right'].set_position(('outward', 60))  # Offset the third axis
ax3.bar(weather_data['Months'], weather_data['Precipitation (mm)'], color='tab:green', alpha=0.3, label='Precipitation')
ax3.set_ylabel('Precipitation (mm)', color='tab:green', fontsize=12)
ax3.tick_params(axis='y', labelcolor='tab:green')

# Combining legends
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
lines3, labels3 = ax3.get_legend_handles_labels()

# Adding grid
ax1.grid(True, linestyle='--', alpha=0.5)

# Adding combined legend outside the plot area
plt.legend(handles=lines + lines2 + lines3, labels=labels + labels2 + labels3, loc='upper right', bbox_to_anchor=(1.3, 1), fontsize=12, title='Variables', frameon=True)
# Adjusting layout
fig.tight_layout()

# Displaying the plot
plt.show()
```
